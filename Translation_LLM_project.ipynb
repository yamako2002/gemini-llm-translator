```python
import google.generativeai as genai
from google.genai import types
import os



genai.configure(api_key='AIzaSyA_Vjl77pm2s19XBdInb1npM4qs422Wbpc') 


def list_available_models():
    print("Available Models:")
    for model in genai.list_models():
        if "generateContent" in model.supported_generation_methods:
            print(f"  - {model.name}")


list_available_models()


model_id = "gemini-flash-latest" 

sys_instruct = "You are a professional word-for-word translator. Output ONLY the translated word."

def translate_word(word, target_lang="Spanish"):
    try:
        response = genai.GenerativeModel(
            model_name=model_id,
            system_instruction=sys_instruct
        ).generate_content(
            f"Translate '{word}' to {target_lang}",
            generation_config={'temperature': 0.1}
        )
        return response.text.strip()
    except Exception as e:
        return f"Error: {e}"


print(translate_word("Apple", "French"))
```
